```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([[i, np.random.random() * 3 + i + 5] for i in range(1, 31)])
```

```python
data
```

```python
plt.scatter(data[:, 0], data[:, 1]);
```

```python
X = np.column_stack((data[:, 0], np.ones(len(data))))
Y = data[:, 1]
```

```python
X
```

```python
Y
```

```python
def func(a0, b0):
    return np.array([np.sum(-2 * (Y - np.dot(X, [a0, b0])) * X[:, 0]), np.sum(-2 * (Y - np.dot(X, [a0, b0])))])
```

```python
w0 = np.array([0, 0])
```

```python
res = []
lam = 0.00001
```

```python
while True:
    res.append(w0)
    w1 = w0 - lam * func(*w0)
    if np.linalg.norm(w1 - w0) <= 0.0001:
        break
    w0 = w1
```

```python
print(w0)
```

```python
plt.plot(data[:, 0], w0[1] + w0[0] * data[:, 0], color='green')
plt.scatter(data[:, 0], data[:, 1]);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(data[:, 0], data[:, 1])

X = data[:, 0].reshape(-1, 1)
Y = data[:, 1]


reg = LinearRegression().fit(X, Y)
print("Coef:", reg.coef_)
print("Intercept:", reg.intercept_)

plt.plot(data[:, 0], reg.predict(X), color='red')
plt.plot(data[:, 0], w0[1] + w0[0] * data[:, 0], color='green');
```
